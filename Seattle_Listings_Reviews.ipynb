```python
import pandas as pd
```

```python
Seattle_listings_df=pd.read_csv("Resources/Seattle_listings.csv")
```

```python
Seattle_reviews1_df=pd.read_csv("Resources/Seattle_reviews1.csv")
Seattle_reviews2_df=pd.read_csv("Resources/Seattle_reviews2.csv",names=Seattle_reviews1_df.columns)
```

```python
Seattle_reviews_df=pd.concat([Seattle_reviews1_df,Seattle_reviews2_df])
Seattle_reviews_df.head()
```

```python
Seattle_listings_df=Seattle_listings_df[["id", "name", "latitude", "longitude",  "property_type", "accommodates", "bathrooms", "bedrooms", "beds", "review_scores_rating"]]
```

```python
Seattle_listings_df=Seattle_listings_df.dropna()
```

```python
Seattle_listings_df.head()
```

```python
bins = pd.cut(Seattle_listings_df['review_scores_rating'], [0,20,40,60,80,100])

Seattle_listings_df.groupby(bins)['review_scores_rating'].count()
```

```python
Seattle_listings_df.rename(columns={'id':'listing_id'}, inplace=True)
Seattle_listings_df.head()
```

```python
Seattle_Listings_Reviews_df=Seattle_listings_df.merge(Seattle_reviews_df, how='outer', on='listing_id')
Seattle_Listings_Reviews_df=Seattle_listings_df.dropna()
Seattle_Listings_Reviews_df["City"]="Seattle"
Seattle_Listings_Reviews_df.head()
```

```python
Seattle_Listings_Reviews_df.to_csv("Output/Seattle_Listings_Reviews_df.csv")
```
