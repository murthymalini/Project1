```python
import pandas as pd
```

```python
Boston_listings_df=pd.read_csv("Resources/Boston_listings.csv")
```

```python
Boston_reviews1_df=pd.read_csv("Resources/Boston_reviews1.csv")
Boston_reviews2_df=pd.read_csv("Resources/Boston_reviews2.csv",names=Boston_reviews1_df.columns)
```

```python
Boston_reviews_df=pd.concat([Boston_reviews1_df,Boston_reviews2_df])
Boston_reviews_df.head()
```

```python
Boston_listings_df=Boston_listings_df[["id", "name", "latitude", "longitude",  "property_type", "accommodates", "bathrooms", "bedrooms", "beds", "review_scores_rating"]]
```

```python
Boston_listings_df=Boston_listings_df.dropna()
```

```python
Boston_listings_df.head()
```

```python
bins = pd.cut(Boston_listings_df['review_scores_rating'], [0,20,40,60,80,100])

Boston_listings_df.groupby(bins)['review_scores_rating'].count()
```

```python
Boston_listings_df.rename(columns={'id':'listing_id'}, inplace=True)
Boston_listings_df.head()
```

```python
Boston_Listings_Reviews_df=Boston_listings_df.merge(Boston_reviews_df, how='outer', on='listing_id')
Boston_Listings_Reviews_df=Boston_listings_df.dropna()
Boston_Listings_Reviews_df["City"]="Boston"
Boston_Listings_Reviews_df.head()
```

```python
Boston_Listings_Reviews_df.to_csv("Output/Boston_Listings_Reviews_df.csv")
```
