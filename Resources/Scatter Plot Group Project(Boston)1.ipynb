```python
Boston_listing_df["Mean Price"] = 0

for index, row in Boston_listing_df.iterrows():
    
    try:
        Boston_listing_df.loc[index,"Mean Price"] = mean_price_by_listing_df.loc[row["id"]]
    except:
        Boston_listing_df.loc[index, "Mean Price"] = 0
        
          
        
        
Boston_listing_df.head()
        

```
