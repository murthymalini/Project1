```python
import pandas as pd
import numpy as np
import os 
import datetime
import matplotlib.pyplot as plt
import scipy.stats as st

```

```python
file1 = "c://Learn//Project1//Resources//Boston_calendar1.xlsx"
boston_df1 = pd.read_excel(file1)
boston_df1.head()
```

```python
import math
boston_df1.describe()
```

```python
boston_df1 = boston_df1.dropna(how="any")
boston_df1.count()
```

```python
file2 = "c://Learn//Project1//Resources//Boston_calendar2.xlsx"
boston_df2 = pd.read_excel(file2)
boston_df2.head()
```

```python
df_final_boston = pd.concat([boston_df1, boston_df2])

df_final_boston.count()
```

```python
df_final_boston.head()
```

```python
df_final_boston['year'] = df_final_boston['date'].dt.year
df_final_boston.head()
```

```python
df_final_boston['Location'] = 'Boston'
df_final_boston.head()
```

```python
df_final_boston.count()
```

```python
df_final_boston.fillna(method = "bfill", inplace = True)
```

```python
df_final_boston['month'] = df_final_boston['date'].dt.month
df_final_boston.head()
```

```python
df_final_boston.tail()
```

```python
df_final_boston["year_and_month"] = df_final_boston["year"].map(str)+ "-" + df_final_boston["month"].map(str)
df_final_boston.head()
```

```python
print("started")
plt.scatter(df_final_boston.iloc[:,7],[df_final_boston.iloc[:,3]])
plt.xlabel('year_and_month')
plt.xticks("year_and_month" rotation =90)
plt.ylabel('price')
plt.title("year_and_month vs price")
plt.show()
print("finished")
```

```python
year_and_month= df_final_boston.iloc[:,7]
price = df_final_boston.iloc[:,3]
corelation = st.pearsonr(year_and_month, price)
print(f"The corelation between both factor is {round(corelation[0],2)}")
```

```python
plt.scatter(df_final_boston["year_and_month"],[df_final_boston["price"]])
plt.xlabel('year_and_month')
plt.ylabel('price')
print(f"The corelation between both factor is {round(corelation[0],2)}")
plt.show()
```
