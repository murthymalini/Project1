```python
import pandas as pd
import numpy as np
import os 
import datetime
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
file1 = "c://Learn//Project1//Resources//Boston_calendar1.xlsx"
seattle_df1 = pd.read_excel(file1)
seattle_df1.head()
```

```python
seattle_df1 = seattle_df1.dropna(how="any")
seattle_df1.count()
```

```python
file2 = "c://Learn//Project1//Resources//Boston_calendar1.xlsx"
seattle_df2 = pd.read_excel(file2)
seattle_df2.head()
```

```python
seattle_df2 = seattle_df2.dropna(how="any")
seattle_df2.count()
```

```python
df_final_seattle = pd.concat([seattle_df1, seattle_df2])

df_final_seattle.count()
```

```python
df_final_seattle.head()
```

```python
df_final_seattle['year'] = df_final_seattle['date'].dt.year
df_final_seattle.head()
```

```python
df_final_seattle['Location'] = 'Seattle'
df_final_seattle.head()
```

```python
df_final_seattle.count()
```

```python
df_final_seattle[["price","year"]]
```

```python
df_final_seattle['month'] = df_final_seattle['date'].dt.month
df_final_seattle.head()
```

```python
df_final_seattle.tail()
```

```python
df_final_seattle["year_and_month"] = df_final_seattle["year"].map(str)+ "-" + df_final_seattle["month"].map(str)
df_final_seattle.head()
```

```python
plt.scatter(df_final_seattle.iloc[:,7],[df_final_seattle.iloc[:,3]])
plt.xlabel("year_and_month")
plt.ylabel("price")
plt.title("year vs price")
plt.xticks("year_and_month" rotation =90)
plt.show()
```
